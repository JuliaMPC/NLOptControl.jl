```julia
using NLOptControl
using Polynomials
using Plots
pyplot()
default(guidefont = font(17), tickfont = font(15), legendfont = font(12), titlefont = font(20))
```

```julia
# for the first test, we will approximate a polynomial
x=-1; # TODO eventually allow for other intervals
x= 1;
x = linspace(x,x,100);
 =  3;
 = -3;
 = -8;
 =  7;
 = Poly([,,,,,,,]);
y = polyval(,x)

# construct polynomial approximation
N = 2 # N = 2 can exactly represent a 3rd order polynomial
if N < 2 error("Please increase N to at least 2!") end
 = LGL_nodes(N);
 = LGL_weights();
D = LGL_Dmatrix();
# approximate the integral
 = polyint();
Y = polyval(,x[end]) - polyval(,x[1]);
C = Y - polyval(,x[end]); # find the constant of integration
y = polyval(,x) + C;
f = polyval(,)
 = cumsum(.*f,1)

# plot the result
plot(x,y,label=string(),w=4)
plot!(x,y,label=@sprintf("integral = %0.3f",y[end]),w=4)
plot!(,,label=@sprintf("approx. integral = %0.3f",[end]),w=4, size=(800,800))
scatter!(,f,label=string("tau with N = ",N),markershape = :hexagon, markersize=10)
#savefig("test2.png")
```

```julia
savefig("test2a.png")
```
